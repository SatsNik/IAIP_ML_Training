```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("taxi.csv")
dataset.shape
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
#seperating dependent-independent variable
x=dataset.iloc[:,:4].values
x
```

```python
y=dataset.iloc[:,4].values
y
```

```python
#split dataset into training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
x_test
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
import pickle #
```

```python
pickle.dump(reg,open('taxi.pkl','wb')) #'wb'=write binary
```

```python
#Test the model
model=pickle.load(open('taxi.pkl','rb')) #'rb'=read binary
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print(r2*100) #r2*100 % accuracy
```

```python
sample_prediction = model.predict([[16,1678000,6500,60]]).round(3) # predict method takes either a datafarame or a 2D numpy array(like 2 dimensional data)
print(sample_prediction)
```
