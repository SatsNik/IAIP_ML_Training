```python
# pivot function is used to reshape
import pandas as pd
d1=pd.DataFrame({'days':[1,2,3,4,5,6],'st_name':['a','b','a','b','a','b'],'eng':[10,12,14,15,16,12],'math':[17,18,19,13,14,16]})
d1.pivot(index='days',columns='st_name')
```

```python
# Group by in pandas
import pandas as pd
d1=pd.DataFrame({'Name':['a','b','c','d','a','b','c'],'s1':[12,13,14,15,12,13,14],'s2':[22,23,24,25,22,23,24]})
d2=d1.groupby('Name')
d2.get_group('a')
```

```python
# Group by in pandas
import pandas as pd
d1=pd.DataFrame({'Name':['a','b','c','d','a','b','c'],'s1':[12,13,14,15,12,13,14],'s2':[22,23,24,25,22,23,24]})
d2=d1.groupby('Name')
for x,y in d2:
    print(x)
    print(y)
```
